```python
# Step 1: Import Libraries
import pandas as pd
import numpy as np
import re
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report, accuracy_score
from nltk.corpus import stopwords
import nltk

# Step 2: Download Stopwords
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

# Step 3: Load Dataset
df = pd.read_csv(r"F:\project.py\data set predection\Sentiment Analysis\Reviews.csv")  # âœ… Change path as needed

# Step 4: Preprocess Text (Using re instead of nltk)
def clean_text(text):
    tokens = re.findall(r'\b\w+\b', str(text).lower())  # Tokenize using regex
    words = [word for word in tokens if word not in stop_words]
    return ' '.join(words)

df['Clean_Review'] = df['Text'].apply(clean_text)

# Step 5: Create Sentiment Labels from Score
def label_sentiment(score):
    if score >= 4:
        return 1  # Positive
    elif score == 3:
        return 0  # Neutral
    else:
        return -1  # Negative

df['Label'] = df['Score'].apply(label_sentiment)

# Step 6: Split Data
X_train, X_test, y_train, y_test = train_test_split(
    df['Clean_Review'], df['Label'], test_size=0.3, random_state=42
)

# Step 7: TF-IDF Vectorization
vectorizer = TfidfVectorizer()
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)

# Step 8: Train the Naive Bayes Model
model = MultinomialNB()
model.fit(X_train_vec, y_train)

# Step 9: Evaluate Model
y_pred = model.predict(X_test_vec)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))

# Step 10: Predict a Custom Review
def predict_sentiment(review):
    cleaned = clean_text(review)
    vectorized = vectorizer.transform([cleaned])
    pred = model.predict(vectorized)[0]
    sentiment_map = {1: "Positive", 0: "Neutral", -1: "Negative"}
    return sentiment_map[pred]

# Example Prediction
custom_review = "food is bade"

print("\nCustom Review Sentiment:", predict_sentiment(custom_review))
```
